```python
thisList=["apple", "banana", "cherry"]
thisList[1]=["blackcurrant", "watermelon", "melon"]
print(thisList)
```

```python
fruits = ["apple", "banana", "cherry", "banana", "kiwi", "mango"]
newlist = [x if x != "banana" else "orange" for x in fruits]
print(newlist)
newlist.sort()
print(newlist)
newlist.sort(reverse=True)
print(newlist)
```

```python
def myfunc(n):
  return abs(n - 50)

thislist = [100, 50, 65, 82, 23]
thislist.sort(key = myfunc)
print(thislist)
```

```python
list1 = ["a", "b" , "c"]
list2 = [1, 2, 3]

for x in list2:
  list1.append(x)

print(list1)
```

```python
list1 = ["a", "b" , "c"]
list2 = [1, 2, 3]

list1.extend(list2)
print(list1)
```
