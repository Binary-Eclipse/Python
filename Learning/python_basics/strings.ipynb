Strings are Arrays
Like many other popular programming languages, strings in Python are arrays of unicode characters.
However, Python does not have a character data type, a single character is simply a string with a length of 1.

Square brackets can be used to access elements of the string.

```python
a="i am not a perfect lover"
print(a[0])
print(a[1])
print(a[2])
```

Check String
To check if a certain phrase or character is present in a string, we can use the keyword in.

```python
a="big apple is not matter of big size, but it is matter of big taste"
print("taste" in a) #boolean value return kore
print(" " in a)
print("mangoes" in a)
print("bigs" not in a)
```

Slicing

You can return a range of characters by using the slice syntax.

Specify the start index and the end index, separated by a colon, to return a part of the string.

positive indexing start from "0 to end-1"
Negative indexing start from end side with "-1" to "-end"

```python
a="apple,banana,mangoes"
print(a[0:5])
print(a[-6:-1])#string[start : end]
print(a[0:5:2]) #string[start : end : step]
```

F-Strings
F-String was introduced in Python 3.6, and is now the preferred way of formatting strings.

To specify a string as an f-string, simply put an f in front of the string literal, and add curly brackets {} as placeholders for variables and other operations.

```python
age=20
txt=f"i am {age} years old"
print(txt)
```

Display the price with 2 decimals:

```python
price=45000
p=f"the shirt price is {price:.2f} tk"
print(p)
```

Perform a math operation in the placeholder, and return the result:

```python
p=f"the shirt container price is {900*10000} tk"
print(p)
```

Escape Character
To insert characters that are illegal in a string, use an escape character.

An escape character is a backslash \ followed by the character you want to insert.

An example of an illegal character is a double quote inside a string that is surrounded by double quotes:

```python
#You will get an error if you use double quotes inside a string that is surrounded by double quotes
#print("We are the so-called "Vikings" from the north.")#error
#But you can use double quotes if you use single quotes to surround the string:
print('We are the so-called "Vikings" from the north.')
#The escape character \ allows you to use double quotes when you normally would not be allowed:
print("We are the so-called \"Vikings\" from the north.")
```
