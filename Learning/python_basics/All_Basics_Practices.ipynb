Global Variables vs Local Variables
### 🔹 Global Variable

A **global variable** is a variable declared outside all functions. It can be accessed and used by any function in the program. Its lifetime lasts for the entire execution of the program. Global variables are useful when multiple functions need to share the same data, but they should be used carefully because they can be modified from anywhere.

---

### 🔹 Local Variable

A **local variable** is a variable declared inside a function or a block. It can only be accessed within that specific function or block where it is defined. Its lifetime is limited to the execution of that function. Local variables are safer and help avoid unintended changes in the program.


```python
name="shakil" #global variable

def myfunc():
    print(name)
myfunc()

def anotherfunc():
    age=25
    
    print("my name is",name,"and I am",age,"years old")
anotherfunc()

# or
#note: all function defined in same order  
def oneMorefunc():
      global lastname
      lastname="patwary" 
oneMorefunc()
print("my name is",name,lastname)
#when i defined a variable inside oneMorefunc() as a global variable, it can be used outside the function as well.
```

Built-in Data Types
Python has the following data types built-in by default, in these categories:

Text Type:	str
Numeric Types:	int, float, complex
Sequence Types:	list, tuple, range
Mapping Type:	dict
Set Types:	set, frozenset
Boolean Type:	bool
Binary Types:	bytes, bytearray, memoryview
None Type:	NoneType

```python
#string
z="Hello World"
print(z)
#int
x=1
print(x)

#float
p=3.14
print(p)

#complex a+bj
r=3+5j
print(r.real)
print(r.imag)

#list
mylist=["apple", "banana", "cherry"]
print(mylist)

#tuple
mytuple=("apple", "banana", "cherry")
print(mytuple)
print(mytuple[0])

#range
y=range(8)
print(y)
print(list(y))

#dict(mapping):A dictionary (dict) in Python is a data type used to store data in key–value pairs. 
# Each key is unique and is used to access its corresponding value.
mydict={"name":"shakil", "age":25, "city":"dhaka"}
print(mydict)
print(mydict["name"])
print(mydict["age"])

# set: A set is a collection which is unordered, unchangeable*, and unindexed.
myset={"apple", "banana", "cherry"}
print(myset)

#frozenset: A frozenset is a collection which is unordered, unchangeable, and unindexed.
fs=frozenset(["apple", "banana", "cherry"])
print(fs)

#bool
bol=True
print(bol)
bool(5 > 2) #automatically returns True because 5 is greater than 2

#bytes
b = bytes([65, 66, 67])
print(b)

b = bytes([65, 66, 67])
print(b[0])   # Output: 65

B=b"Hello"
print(B[0])

#bytearray
ba=bytearray([65, 66, 67])
print(ba)

#memoryview
mv=memoryview(bytes(5))
print(mv)

#none
x=None
print(x)
```

str() - constructs a string from a wide variety of data types, including strings, integer literals and float literals
