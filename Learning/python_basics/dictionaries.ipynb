Dictionaries Syntax : 
dict={ key : value }

its changeable,ordered,
not allow duplicate key

#not allowed key
d = {
    [1, 2]: "list key",   # ❌
    {1, 2}: "set key"     # ❌
}

#allowed key            
d = {
    1: "int key",
    3.14: "float key",
    "name": "string key",
    (1, 2): "tuple key"
} 
can store any types of datatype(int,float,string,bool,list,tuple,set)

```python
mydict={
    "name": "shawon",
    "age": 15,
    "height": 5.6,
    "result": {"math": 80, "english": 90, "physics": 85},
    "student": True,
    "sets": {1, 2, 3, 4, 5},
    "lists": [1, 2, 3, 4, 5],
    "tuples": (1, 2, 3, 4, 5)
}

print(mydict)
#accessing values
print(mydict["name"])
print(mydict["age"])
print(mydict["height"])
print(mydict["result"])
print(mydict["student"])
print(mydict["sets"])
print(mydict["lists"])
print(mydict["tuples"])

#accesing values using get method
x=mydict.get("name")
print(x)
```

```python
#getting all keys
print(mydict.keys()) # return as list 
#getting all values
print(mydict.values())
```

```python
#modified value
mydict["name"]="shakil"
print(mydict["name"])

#adding new key value pair
mydict["country"]="Bangladesh"
print(mydict)

#update
mydict.update({"name": "shawon", "age":16})
print(mydict)
```

The items() method will return each item in a dictionary, as tuples in a list.

```python
x=mydict.items()
print(x)
```

Check if Key Exists

```python
if "name" in mydict:
    print("name is present in the dictionary")
```


🧠 Why you can’t change keys directly

Dictionary keys must be:

Immutable
Hashable

If Python allowed changing keys, it would break the hashing system (same concept as sets).

Correct way to change a key
Method 1: Using pop()

```python
d={"name":"shawon"}
d["fullname"]=d.pop("name")
print(d)
```

Removing Items
pop()
popitem()  #The popitem() method removes the last inserted item
del        #The del keyword can also delete the dictionary completely
clear()    #The clear() method empties the dictionary:

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
#pop
thisdict.pop("model")
print(thisdict)

#add new key value pair
thisdict["color"]="red"
print(thisdict)

#popitem()
thisdict.popitem() 
print(thisdict)

#clear
thisdict.clear()
print(thisdict)
```

Loop:

#Print all key names in the dictionary, one by one:
#Print all values in the dictionary, one by one:
#also use the values() method to return values of a dictionary:
#use the keys() method to return the keys of a dictionary:
#Loop through both keys and values, by using the items() method:




```python
#Print all key names in the dictionary, one by one:
for x in mydict:
    print(x)

#Print all values in the dictionary, one by one:
for x in mydict:
    print(mydict[x])

#also use the values() method to return values of a dictionary:
for x in mydict.values():
    print(x)

#use the keys() method to return the keys of a dictionary:
for x in mydict.keys():
    print(x)

#use the items() method to return the key:value pairs of a dictionary, as tuples in a list:
for x in mydict.items():
    print(x)
```

```python
#nested dictionary

myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}
print(myfamily)

#accessing nested dictionary
print(myfamily["child1"]["name"])


#loop through nested dictionary
for x, obj in myfamily.items():
  print(x)

  for y in obj:
    print(y + ':', obj[y])
```
