2.Reverse a string

```python
st="Hello"
#using slicing
print(st[::-1])

#using loop
rev=""
for ch in st:
    rev=ch+rev
print(rev)

#using recursion
def reverse_string(st):
    if len(st) == 0:
        return st
    else:
        return st[-1] + reverse_string(st[:-1])
    
print(reverse_string(st))

 #using built-in function
print(''.join(reversed(st)))
```

6.Find Frequency of character

```python
st="Hello"
print(len(st))

#using dictionary
freq={}
for x in st:
    if x in freq:
        freq[x]+=1
    else:
        freq[x]=1

print(freq)

#frequency count another way
freq = {}
for ch in "Hello":
    freq[ch] = freq.get(ch, 0) + 1
#using pretty format
for k, v in freq.items():
    print(k, "=", v)
```

7. Remove duplicates from the list

10.find max/min value in list

```python
#using max() & min()
lst=[1,5,4,3,2]
print("Max:", max(lst))
print("Min:", min(lst))

#using loop
max_value=lst[0]
min_value=lst[0]
for i in lst:
    if i>max_value:
        max_value=i
    if i<min_value:
        min_value=i

print("Max:", max_value)
print("Min:", min_value)


#using sort
lst.sort()

print("Max:", lst[-1])
print("Min:", lst[0])
```

11. reverse a list

```python
lst=[1,2,3,4,5]
#using slicing
print(lst[::-1])

#using reverse() inplace
lst.reverse()
print(lst) # change the orginal list

#using reversed()
print(list(reversed(lst)))

#using loop
rev=[]
for x in lst:
    rev.insert(0,x) # insert at index 0, so it will be in reverse order(adds at beginning)
 
print(rev)
#adds at end
lst.append([9, 6])
print(lst)
```

12.merge two dictionaries

```python
dict1={"name":"Alice", "age":30, "city":"New York"}
dict2={"name":"Bob", "age":25, "city":"Los Angeles"}

#using update() method
dict1.update(dict2)
print(dict1)

#Using unpacking (clean way)
merged_dict = {**dict1, **dict2}
print(merged_dict)

#using loop
for key, value in dict2.items():
    dict1[key] = value

print(dict1)
```
