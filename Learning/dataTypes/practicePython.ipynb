# list

```python
#declaration
mylist=["apple","banana","cherry"]
print(mylist)

#length 
print(len(mylist))

#list type
print(type(mylist))

#list contain multiple data types
mylist1=["apple",1,True,3.5,[1,2,30],("a"),{"name":"shakil"}]
print(mylist1)

#change make in list
mylist[1]="orange"
print(mylist)

mylist[3:5]=["kiwi","mango"]
print(mylist)

#insert new item in list
mylist.insert(5,"grapes")
print(mylist)

#append
mylist.append("watermelon")

#extend
mylist.extend(["papaya","guava"])
print(mylist)

digit=[1,2,3,4,5]
mylist.extend(digit)
print(mylist)


#remove
mylist.remove("kiwi")
print(mylist)

mylist.pop(1)
print(mylist)

del mylist[0]
#del mylist
print(mylist)

mylist.clear()
```

# Tuple

```python
#declaration
tpl=('a', 'b', 'c', 'd', 'e')
print(tpl)

#length
print(len(tpl))

#type
print(type(tpl))

#cannot apply change,insert,extend and append in tuple

#change can be done by converting tuple to list and then back to tuple
list1=list(tpl)
list1.remove('c')
list1.append('f')
list1.insert(2,'g')
list1.extend(['h','i'])
tpl=tuple(list1)
print(tpl)
#or can apply change in tuple by using tuple concatenation
tpl=tpl+('j',)
print(tpl)
#or can apply change in tuple by using tuple slicing
tpl=tpl[:3]+['k']+tpl[3:]
print(tpl)
#or can apply change in tuple by using tuple unpacking
a,b,c,d,e,f,g,h,i,j,k=tpl
j='a'
print((a,b,c,d,e,f,g,h,i,j,k))
#or can apply change in tuple by using tuple methods
  #tuple onl have two methods count() and index()
      print(tpl.count('a'))
      print(tpl.index('a'))
#or can apply change in tuple by using tuple comprehension
 tpl=tuple(x for x in tpl if x != 'a')
print(tpl)
#or can apply change in tuple by using tuple functions
new_tpl=tuple(map(lambda x: x.upper(), tpl))
print(new_tpl)
new_tple=tuple(filter(lambda x: x != 'a', tpl))
print(new_tple)
new_tpl=tuple(sorted(tpl))
print(new_tpl)
newtpl=tuple(reversed(tpl))
print(newtpl)
#or can apply change in tuple by using tuple operations
t1=(1,2,3)
t2=(4,5,6)
t3=t1+t2
print(t3)
```

# set

```python
sets={1,2,3,4,5}
#add
sets.add(6)
print(sets)

#remove
sets.remove(3)
print(sets)
#copy
y=sets.copy()
print(y)

#difference
print(sets.difference(y))

#discard
sets.discard(2)
print(sets)

#difference_update
sets.difference_update(y)
print(sets)

#after update difference
print(len(sets)) # 0

sets1={1,2,3,4,5}
sets2={1,5,4,7,5}

#intersection
print(sets1.intersection(sets2))

#intersecion_update
print(sets1.intersection_update(sets2))#its often show none because it update the sets1 and sets1 become empty set
print(sets1)

#issubset
print(sets1.issubset(sets2))

#issuperset
print(sets1.issuperset(sets2))

#symmetric_difference
print(sets1.symmetric_difference(sets2))

#union
print(sets1.union(sets2))

#update
print(sets1.update(sets2))#show none because it update the sets1 and sets1 become {1,2,3,4,5,7}
sets1.update(sets2)
print(sets1)

#isdisjoint
print(sets1.isdisjoint(sets2))

#pop
print(sets1.pop())#return the first element of the set and remove it from the set

#frozenset
k=frozenset([1,2,3,4,5])
print(k)

#testing membership
s={1,2,2,3}
print(2 in s)  # True
print(4 in s)  # False

#removing duplicates from a list using set
unique=set(s)
print(unique)
```

# Casting
means converting a value from one datatype to another

```python
s=str(3)
print(type(s))

#string to int
i=int(s)
print(type(i))

#int to complex
c=complex(i)
print(type(c))

#int to float
f=float(i)
print(type(f))

#float to int
ii=int(f)
print(type(ii))

#complex to int 
#just real part of complex number can be converted to int
iii=int(c.real)
print(type(iii))

#extract real and imaginary part of complex number
print("Real part:", c.real)
print("Imaginary part:", c.imag)

list1=[1,2,3,4,5]

#list to tuple
t=tuple(list1)
print(type(t))

#tuple to list
l=list(t)
print(type(l))

#list to set
s=set(list1)
print(type(s))

#set to list
L1=list(s)
print(type(L1))

#set to tuple
T1=tuple(s)
print(type(T1))

#tuple to set
S1=set(T1)
print(type(S1))

#list to Dictionary
dict1={i: list1[i] for i in range(len(list1))}
print(type(dict1))
```
