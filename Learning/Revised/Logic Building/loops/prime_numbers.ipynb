# 4. 🔢 Prime Numbers

A prime number has exactly two factors:

```text
1 and itself
```

Example:

```text
2, 3, 5, 7, 11, 13
```

## Basic

1. Check whether a number is prime.
2. Print primes from `1` to `n`.
3. Count primes in a range.
4. Find the first `n` prime numbers.
5. Find the next prime number.
6. Find the previous prime number.

## Intermediate

7. Find prime factors.
8. Find the largest prime factor.
9. Find the sum of prime numbers.
10. Find two consecutive primes.
11. Check whether two numbers are twin primes.
12. Check whether a number is a circular prime.

## Advanced

13. Sieve of Eratosthenes.
14. Find prime factorization.
15. Find distinct prime factors.
16. Find numbers with exactly `n` factors.
17. Find the `n`th prime number.
18. Find prime numbers in a large range.
