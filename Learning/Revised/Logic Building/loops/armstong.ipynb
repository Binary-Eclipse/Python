# 5. 💪 Armstrong Numbers

Example:

```text
153 = 1³ + 5³ + 3³
```

Therefore, `153` is an Armstrong number.

## Basic

1. Check whether a number is Armstrong.
2. Print Armstrong numbers from `1` to `n`.
3. Count Armstrong numbers in a range.

## Intermediate

4. Handle numbers with different digit lengths.
5. Find all Armstrong numbers between two numbers.
6. Calculate the Armstrong sum manually.
7. Check Armstrong numbers using a function.

## Advanced

8. Find Armstrong numbers with `n` digits.
9. Find the largest Armstrong number below a limit.
10. Optimize the calculation for large ranges.
