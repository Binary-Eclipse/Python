# 7. 🔁 Palindrome Numbers

A palindrome reads the same forward and backward.

```text
121
```

Reverse:

```text
121
```

## Basic

1. Check whether a number is palindrome.
2. Check whether a string is palindrome.
3. Print palindrome numbers in a range.
4. Count palindrome numbers.

## Intermediate

5. Find the next palindrome.
6. Find the previous palindrome.
7. Find the largest palindrome below `n`.
8. Check palindrome without converting to string.

## Advanced

9. Find the longest palindromic number in a list.
10. Check palindrome using recursion.
11. Generate palindromic numbers.
