# 3. 🐇 Fibonacci Series

The Fibonacc sequence:

```text
0, 1, 1, 2, 3, 5, 8, 13...
```

## Basic

1. Print the first `ni` Fibonacci numbers.
2. Print Fibonacci numbers using `while`.
3. Find the `n`th Fibonacci number.
4. Check whether a number belongs to Fibonacci.
5. Find the sum of the first `n` Fibonacci numbers.

## Intermediate

6. Generate Fibonacci using a function.
7. Generate Fibonacci using recursion.
8. Generate Fibonacci using a generator.
9. Print Fibonacci numbers within a range.
10. Print only even Fibonacci numbers.
11. Print only odd Fibonacci numbers.

## Advanced

12. Compare iterative and recursive Fibonacci.
13. Find Fibonacci using memoization.
14. Find Fibonacci using dynamic programming.
15. Find the largest Fibonacci number below `n`.
16. Find the first Fibonacci number containing `k` digits.
