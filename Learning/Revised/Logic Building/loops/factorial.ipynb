## find the factorial of all list elements

````python
#case1: 0! or 1! both return 1
#case2: if x<0 then reject
def factorial_gen(x):
  if x<0 :
    print("Rejected")
  else:     
      if x==0 or x==1:
            yield 1
      else:
         yield x*factorial_gen(x-1)

lst=[3,6,1,0]    
fact_lst=[]

for i in lst:
    fact=factorial_gen(i)
    fact_lst.append(fact)

largest=max(fact_lst)
print(largest)
````

## trailing 0

````python
#case1: 0! or 1! both return 1
#case2: if n<0 then reject
def factorial_gen(x):
  if x<0 :
    return "rejected"
  else:     
      if x==0 or x==1:
            yield 1
      else:
         yield x*factorial_gen(x-1)

def trailing_zero(f):
   count=0
   while f >= 5:
      f=f//5
      count+=f
   yield count

f=factorial_gen(125)
print(trailing_zero(f))

````

## number of digit in n!

````python
def factorial_gen(x):
  if x<0 :
    return "rejected"
  else:     
      if x==0 or x==1:
            yield 1
      else:
         yield x*factorial_gen(x-1)

def find_num_of_digit(result):
   c=0
   if result == 0:
      c=1
   else:
      result=abs(result)
      while result > 0:
         result//=10
         c+=1
   return c

r=factorial_gen(10)
print(find_num_of_digit(r))
````

## find last non-zero digit

````python
def factorial_gen(x):
  if x<0 :
    return "rejected"
  else:     
      if x==0 or x==1:
            yield 1
      else:
         yield x*factorial_gen(x-1)

def find_non_zero_digit(res):
   
   while res%10==0:
      res //=10

   return res%10

r=factorial_gen(10)
print(find_non_zero_digit(r))
````
