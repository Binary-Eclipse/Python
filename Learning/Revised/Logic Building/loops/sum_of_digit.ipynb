# 9. ➕ Sum of Digits

Example:

```text
12345
```

```text
1 + 2 + 3 + 4 + 5 = 15
```

## Problems

1. Find the sum of digits.
2. Find the sum using `while`.
3. Find the sum using a function.
4. Find the sum of even digits.
5. Find the sum of odd digits.
6. Find the sum of digits at even positions.
7. Find the sum of digits at odd positions.
8. Repeatedly sum digits until one digit remains.

Example:

```text
9875
9 + 8 + 7 + 5 = 29
2 + 9 = 11
1 + 1 = 2
```
