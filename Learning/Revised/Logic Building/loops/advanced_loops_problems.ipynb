## Loop Optimization

11. Compare `for` and `while`.
12. Reduce unnecessary nested loops.
13. Stop early using `break`.
14. Use `continue` efficiently.
15. Compare `O(n)` and `O(n²)` solutions.
16. Use mathematical shortcuts instead of loops.

Example:

Instead of:

```python
total = 0

for i in range(1, n + 1):
    total += i
```

Mathematical approach:

[
\frac{n(n+1)}{2}
]

The loop is:

```text
O(n)
```

The formula is:

```text
O(1)
```

This is an important transition from **basic programming to algorithmic thinking**.
