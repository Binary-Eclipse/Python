# Basic Loop

```python
#Beginner

n=int(input())
#print 1 to n
for i in range (1,n+1):
   print(i)

# print n to 1
for i in range(n+1,1):
   print(i)

#print all even numbers
for i in range(1,n+1):
   if i%2==0:
      print(i)

# print all odd numbers
for i in range(1,n+1):
   if i%2 != 0:
      print(i)

#multiples of 5
for i in range(1,n+1):
   print(i*5)






```

```python
# divisible by 3
for i in range(1,n+1):
   if i%3==0:
      print(i)
```

```python
#numbers between two numbers
firstnumber=int(input())
secondnumber=int(input())

for i in range(firstnumber+1,secondnumber):
   print(i)
```

```python
#sum of 1 to n
n=10
s=0
for i in range(1,n+1):
   s+=i

print(s)
```

```python
#count even numbr in a range
c=0
for i in range(1,59):
    if i%2==0:
        c+=1

print(c)
```

```python
#find the largest number in a range
mxm=1
n=9
for i in range(2,n+1):
    if i > mxm:
        mxm=i
print(mxm)

lst=[1,5,8,9,4,5]
largest=max(lst)
print(largest)
```

```python
#intermediate
#Print numbers divisible by both `3` and `5`.
lst=[2,6,7,9,12]
c=0
for i in lst:
    if i%3==0 and i%5==0 :
        print(i)
        c+=1

if c==0:
 print("None")
```

```python
#Print numbers not divisible by `3`.
lst=[2,6,7,9,12]
c=0
for i in lst:
    if i%3!=0:
        print(i)
        c+=1
if c==0:
    print("None")
```

## find the average of numbers 1 to n

```python
import random
n=random.randint(1,78)
s=0;
for i in range(1,n+1):
    s+=i
print(f"{s/n:.2f}")
```

## print numbers in reversorder

```python
p=7
#way 1
for i in range(p,0,-1):  #-1 = decreased by one
    print(i)

#way2
for i in reversed(range(1,p+1)):
    print(i)

#way3 
while p>=1:
    print(p)
    p-=1

#if way 3 placed in position of way 2 so present way2 at position way3 not will be work becasue p=0 stand after decreased in while loop
```

## print divided by 3

```python
b=9
for i in range(1,b+1):
    if i%3==0:
       continue
    else:
       print(i)
```

## print not divided by 3

```python
b=9
for i  in range(1,b+1):
    if i%3==0:
        break
    else:
        print(i)
```
