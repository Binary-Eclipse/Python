# 6. ✨ Perfect Numbers

A perfect number equals the sum of its proper divisors.

Example:

```text
6 = 1 + 2 + 3
```

Therefore:

```text
6 is a perfect number
```

## Basic

1. Check whether a number is perfect.
2. Print perfect numbers up to `n`.
3. Count perfect numbers in a range.

## Intermediate

4. Find all divisors.
5. Find the sum of proper divisors.
6. Find abundant numbers.
