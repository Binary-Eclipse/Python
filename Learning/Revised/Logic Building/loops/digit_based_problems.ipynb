These are especially important for `while` loops.

1. Count digits.
2. Sum digits.
3. Product of digits.
4. Reverse number.
5. First digit.
6. Last digit.
7. Largest digit.
8. Smallest digit.
9. Count even digits.
10. Count odd digits.
11. Count zeroes.
12. Remove a digit.
13. Replace a digit.
14. Find repeated digits.
15. Find unique digits.
16. Check digits are increasing.
17. Check digits are decreasing.
18. Check digits are alternating.
19. Find the frequency of each digit.

Example:

```text
122333
```

```text
1 → 1 time
2 → 2 times
3 → 3 times
```
