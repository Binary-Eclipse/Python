# 11. 📊 Multiplication Tables

## Basic

1. Print the table of a number.

```text
5 × 1 = 5
5 × 2 = 10
...
```

2. Print tables from `1` to `10`.
3. Print tables within a range.

## Intermediate

4. Print a multiplication table matrix.

```text
1   2   3
2   4   6
3   6   9
```

5. Print only even multiplication results.
6. Print only multiples of `5`.

## Advanced

7. Create a multiplication quiz.
8. Generate random multiplication questions.
9. Calculate score and accuracy.
