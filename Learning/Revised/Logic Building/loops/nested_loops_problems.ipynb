## Important Practice

1. Multiplication table matrix.
2. All pairs from a list.
3. All combinations of two numbers.
4. All combinations of three numbers.
5. Duplicate detection.
6. Matrix traversal.
7. Matrix addition.
8. Matrix multiplication.
9. Pattern printing.
10. Coordinate generation.

Example:

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

Output conceptually:

```text
1 1
1 2
1 3
2 1
2 2
2 3
3 1
3 2
3 3
```
