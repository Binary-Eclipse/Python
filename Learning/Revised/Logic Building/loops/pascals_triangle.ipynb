# 15. 🔺 Pascal's Triangle

Example:

```text
        1
       1 1
      1 2 1
     1 3 3 1
    1 4 6 4 1
```

## Basic

1. Print Pascal's Triangle.
2. Generate rows using lists.
3. Generate rows using loops.

## Intermediate

4. Calculate each value using:

[
C(n,r) = \frac{n!}{r!(n-r)!}
]

5. Find the sum of each row.
6. Find the largest value in a row.

## Advanced

7. Generate Pascal's Triangle without factorial.
8. Find a specific value.
9. Check whether a number exists in Pascal's Triangle.
10. Use Pascal's Triangle to expand binomial expressions.
