```python
# positive/Negative
x=input("Enter the value: ")

if x > 0 :
    print("X is positive")
else: 
    print("X is negative")
```

```python
# Odd/Even
x=input("Enter the value: ")

if x % 2 == 0 :
    print("X is Even")
else: 
    print("X is Odd")
```

```python
# largest of 2
x,y=input("Enter x & y : ")

if x > y :
      print(f"{x} is largest than {y}")
elif x==y:
      print(f"{x},{y} both are equal")
else: 
      print(f"{y} is largest than {x}")
```

```python
# largest of 3
x,y,z=input("Enter x , y & z : ").split()

if x == y == z :
    print(f"{x},{y},{z} are equal")

elif x==y :
     if x > z :
         print(f"{x},{y} equal and largest than {z}")
     else: 
         print(f"{z} is largest than {x} and {y}")

elif y==z :
     if z > x :
         print(f"{z},{y} equal and largest than {x}")
     else: 
         print(f"{x} is largest than {z} and {y}")

elif z==x :
     if x > y :
         print(f"{x},{z} equal and largest than {y}")
     else: 
         print(f"{y} is largest than {x} and {z}")

elif x > y and  x > z:
    print(f"{x} is largest than {y} & {z}")
elif y > x and y>z :
    print(f"{y} is largest than {x} & {z}")
elif z > x and z >y :
    print(f"{z} is largest than {x} & {y}")


```

```python
# smallest of 3
x,y,z=input("Enter x , y & z : ").split()

if x == y == z :
    print(f"{x},{y},{z} are equal")

elif x==y :
     if x < z :
         print(f"{x},{y} equal and smallest than {z}")
     else: 
         print(f"{z} is smallest than {x} and {y}")

elif y==z :
     if z < x :
         print(f"{z},{y} equal and smallest than {x}")
     else: 
         print(f"{x} is smallest than {z} and {y}")

elif z==x :
     if x < y :
         print(f"{x},{z} equal and smallest than {y}")
     else: 
         print(f"{y} is smallest than {x} and {z}")

elif x < y and  x < z:
    print(f"{x} is smallest than {y} & {z}")
elif y < x and y < z :
    print(f"{y} is smallest than {x} & {z}")
elif z < x and z < y :
    print(f"{z} is smallest than {x} & {y}")


```

```python
#Leap Year
year=int(input("Enter year: "))

if year % 400 == 0 :
    print("Leap Year")
elif year % 100 != 0 and year % 4 == 0 :
    print("Leap Year")
else:
    print("Not a Leap Year")
```
