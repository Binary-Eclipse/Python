```python
# largest of 2
x,y=input("Enter x & y : ")

if x > y :
      print(f"{x} is largest than {y}")
elif x==y:
      print(f"{x},{y} both are equal")
else: 
      print(f"{y} is largest than {x}")
```

```python
# largest of 3
x,y,z=input("Enter x , y & z : ").split()

if x == y == z :
    print(f"{x},{y},{z} are equal")

elif x==y :
     if x > z :
         print(f"{x},{y} equal and largest than {z}")
     else: 
         print(f"{z} is largest than {x} and {y}")

elif y==z :
     if z > x :
         print(f"{z},{y} equal and largest than {x}")
     else: 
         print(f"{x} is largest than {z} and {y}")

elif z==x :
     if x > y :
         print(f"{x},{z} equal and largest than {y}")
     else: 
         print(f"{y} is largest than {x} and {z}")

elif x > y and  x > z:
    print(f"{x} is largest than {y} & {z}")
elif y > x and y>z :
    print(f"{y} is largest than {x} & {z}")
elif z > x and z >y :
    print(f"{z} is largest than {x} & {y}")


```

```python
# smallest of 3
x,y,z=input("Enter x , y & z : ").split()

if x == y == z :
    print(f"{x},{y},{z} are equal")

elif x==y :
     if x < z :
         print(f"{x},{y} equal and smallest than {z}")
     else: 
         print(f"{z} is smallest than {x} and {y}")

elif y==z :
     if z < x :
         print(f"{z},{y} equal and smallest than {x}")
     else: 
         print(f"{x} is smallest than {z} and {y}")

elif z==x :
     if x < y :
         print(f"{x},{z} equal and smallest than {y}")
     else: 
         print(f"{y} is smallest than {x} and {z}")

elif x < y and  x < z:
    print(f"{x} is smallest than {y} & {z}")
elif y < x and y < z :
    print(f"{y} is smallest than {x} & {z}")
elif z < x and z < y :
    print(f"{z} is smallest than {x} & {y}")


```

```python
#Leap Year
year=int(input("Enter year: "))

if year % 400 == 0 :
    print("Leap Year")
elif year % 100 != 0 and year % 4 == 0 :
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
name = input("Enter Your Name: ")
age = int(input("Enter Age: "))
weight = float(input("Enter Your Weight (kg): "))
height = float(input("Enter Your Height (e.g. 5.2): "))

print("Name:", name)
print("Age:", age)
print("Weight:", weight)
print("Height:", height)

integer_part = int(height)
decimal_part = height - integer_part

height_convert_into_meter = integer_part * 0.3048 + decimal_part * 0.0254

BMI = weight / (height_convert_into_meter ** 2)

print("BMI =", round(BMI, 2))

if BMI < 18.5:
    print("Underweight")
elif BMI <= 24.9:
    print("Normal")
elif BMI <= 29.9:
    print("Overweight")
elif BMI <= 34.9:
    print("Obesity Class I")
elif BMI <= 39.9:
    print("Obesity Class II")
else:
    print("Obesity Class III")
```

```python
org_password = "321&5"

attempts = 3

while attempts > 0:
    password = input("Enter password: ")

    if len(password) < 5:
        print("Password must be at least 5 characters.")
        continue

    if password == org_password:
        print("Successfully Login")
        break

    attempts -= 1
    print(f"Incorrect Password. Attempts left: {attempts}")

if attempts == 0:
    print("Account Locked!")
```

```python
power=float(input("Enter power(in Watt): "))
hour,minute,second=map(int,input("Enter time(hh:mm:ss): ").split(":"))

totalhour=hour+(minute/60)+(second/(60 ** 2))

unit= (power*totalhour)/1000

if unit <= 100.0 :
    bill=unit*7
elif unit <= 200.0:
    bill=(100*7)+((unit-100)*9)
else:
    bill=(100*7)+(100*9)+((unit-200)*11)

print("Total Bill: ",bill)
```

```python
#tringle type: angle  based
a,b,c=map(float, input("Enter Tringle angles: ").split(" "))

if a+b+c != 180:
    print("Invalid Tringle")
elif a < 90 and b < 90 and c < 90:
    print("Acute Tringle:")
elif a == 90 or b == 90 or c == 90:
    print("Right Tringle")
else:
    print("Obtuse Tringle")
```

```python
# tringle type : side based
a,b,c=map(float, input("Enter Tringle Sides: ").split(" "))

if a==b==c:
    print("Equilateral Tringle")
elif a==b or b==c or c==a:
    print("isoscales Tringle")
else:
    print("Scalene Tringle")
```

```python
#Quadratic Equation
import math

a = float(input("Enter value of a: "))
b = float(input("Enter value of b: "))
c = float(input("Enter value of c: "))


if a == 0:
    print("Not a quadratic equation")

else:
    D = b**2 - 4*a*c

    if D > 0:
        x1 = (-b + math.sqrt(D)) / (2*a)
        x2 = (-b - math.sqrt(D)) / (2*a)

        print("Two real roots")
        print("x1 =", x1)
        print("x2 =", x2)

    elif D == 0:
        x = -b / (2*a)

        print("Equal roots")
        print("x =", x)

    else:
        print("No real roots")
```
