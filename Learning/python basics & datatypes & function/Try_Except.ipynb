# Try_Excepts

 - try block : let's test a block of code for errors
 - except block : let's handle the error
 - else : let's you execute code when there is no error
 - finally : execute code, regardless of the result of the try and except blocks

```python
try:
    print(x)
except:
    print("x is not defined")
else: 
    print("Nothing went wrong")
finally: 
    print("The 'try-except' is finished")
```

# using object can open,edit and close a file

```python
try:
    f=open("mybook.txt")
    try:
        f.write("loreum ipsum")
    except:
        print("something went wrong when writting the file")
    finally:
        f.close()
except:
    print("something went wrong when opening the file")
```
