# JSON 
 - syntax for storing and exchanging data
 - JSON is text written with javascript object notation

# Convert from JSON to Python

```python
import json

#json data
x='{"name":"john","age":30}'

#parse x
y=json.loads(x)

print(y["age"]) # the result is dictionary
```

# python to JSON

```python
import json
x={
    "name":"helal",
    "age": 25
}

y=json.dumps(x)
print(y)
```

# Format the result
json.dumps() has parameters to make it easier to read the result

```python
import json
txt="best for buy"
x=json.dumps(txt,indent=4)
```

# using separtors parameter
              sepearators(",",":")
              separators(".","=")

```python
y=json.dumps(txt,indent=4,separators(",","="))
```

# order the result using 
        sort_keys

```python
z=json.dumps(x,indent=4,sort_keys=True)
```
