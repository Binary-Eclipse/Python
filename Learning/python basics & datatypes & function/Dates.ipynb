# Dates
is not a datatype but can import as datetime to work dates as date object

# current time print

```python
import datetime
print(datetime.datetime.now())
```

# return year and name of weekday 

```python
import datetime
print(datetime.datetime.now().year)
print(datetime.datetime.now().strftime("%A"))
```

# strftime() method take one parameter only
print(datetime.datetime.now().strftime("%y"))

# create date object using datetime() constructor
 - datetime() construtor also take parameter for time and timezone (hour,minute,second,microsecond,tzone) ,but they are optional
 - also has default value of 0 (None for time zone) 00:00:00

```python
import datetime
print(datetime.datetime(2020,5,17))
```
