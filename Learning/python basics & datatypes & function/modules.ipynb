# Modules: same as code library

# create Modules

```python
# save file as: modules.py
def greeting(name):
    print("Hello, "+name)
```

# use module 

```python
import modules
print(modules.greeting("shakil"))
```

# Varibles in modules
 - above greeting function was defined in module
But  modules can contain also varibles like : arrays,tuple,dictionary,object etc

```python
#define a variable dictionary
persion1={
    "name":"helal"
    "age": 25
}


# now can access
import modules
a=modules.persion1["age"]
print(a)
```

# Renaming modules

```python
import modules as mx
print(mx.persion1["age"])# now using mx ,can access the value of age keys from defined dictionary in modules by renaming modules as mx
```

# built in modules
 - has several, lets discussed 2
    1. platform : return platform system features
    2. from: using from we can import a part from a modules

```python
#file name: modules.py
def greeting(name):               # defined funciton in modules
    print("hello!, "+name)
 
persion1={                       # defined variable in modules
    "name"="helal"
    "age"=25
}
#uses show in below in from modules
```

```python
# platform module
import platform
print(platform.system())

# from 
from modules import persion1
print(persion1["age"])

#Note: Don't use modules.persion1["age"] when you import *persion1* variable from *modules*
```
