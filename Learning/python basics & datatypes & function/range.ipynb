# Range
  - Range is a in-built class and it themselves an object
  - its immutable means can't change after creating 
  - it use little memory

```python
# range() with one , two and three arguments
range(10) # start 0(defualt) to stop when 10  / 0-9
range(1,10) # start with 1 and stop when 10 / 1-10
range(1,10,2)# start with 1 and stop when 10 with steps 2 / 1-9 with 1,3,5,7,9

print i in range(1,10,2):
    print (i)
```

# Range is a in-built class and it themselves an object


```python
r= range(10,20,2)
print(r.start)
print(r.stop)
print(r.step)
```

# range is immutable 

```python
# exmaple : it may occurs error
r=range(5)
r[0]=10 # TypeError: 'range' object does not support item assignment
r.append(6) # AttributeError: 'range' object has no attribute append
```

# range uses little memory
By internally stores onle
    - start 
    - stop
    - step
doesnot store all values individually
