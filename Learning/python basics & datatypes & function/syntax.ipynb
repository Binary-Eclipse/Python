# basic print 

```python
#print 
print ("i am failure")

#print text and numbers
print("i am",24,"years old")
```

# print statement with variable

```python
# use f string
age=24
print(f"i am {age} years old")

# use format method
print("i am {} years old".format(age))

# use comma
print("i am",age,"years old")

# use '+'
name="helal"
print ("Hello, "+name) 
print("Hello! , "+name+"Welcome to the world of Python programming.")
```

# print decimal value with take specific number of decimal place

```python
y=3.89675
print(f"the value of y is {y:.2f}")
```

# if statement

```python
if 5 > 2:
    print("Five greater than 2")
else:
    print("Five is not greater than 2")
```
