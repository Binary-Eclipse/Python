# String formatting
  - F-string introduced in python 3.6 before had to use format()

```python
x=890.45678
txt1=f"the value of x : {x}"
txt2=f"the valuse of x : {x:.2f}"
print(txt2)
txt3=f"the valuse of y : {506.67678:.2f}"
print(txt3)

#operation performs
print(f"the multiplication result is: {x*(20*3) : .3f}")

#if..else
txt4=f"it is very {'expensive'if x > 850 else 'cheap'}"
print(txt4)

#execute function
fruits='apple'
txt5=f"i love {fruits.upper()}"

#function
def converter(x):
    return x*0.3048

txt6=f"the plane is flying at a {converter(30000)} meters altitude"
print(txt6)


#comma as thousand separators
p=10200034
txt7=f"the price is {p:,} taka"
print(txt7)


#using format()
print("the price is {}".format(x))
print(f"the price is "+format(x))


#multiple values
n=9
m=0
k=4.6788
txt8="the value of n:{} , m:{} and k:{}"
print(txt8.format(n,m,k))


#indexing
txt8="the value of n:{0} , m:{1} and k:{2:1f}"
print(txt8.format(n,m,k))

#refers to the same vale more than once
tx9="This is a {1} and {1} is used by {0}"
print(tx9.format(n,m))

#name indexes
myOrder="i have a {carname}, it is a {model}"
print(myOrder.format(carname="ford" , model="Mustang"))
```
