# RegEx

```python
import re

# re.match()
text="python is easy"
rs=re.match(r"python",text)
if rs:
    print(rs.group())

#re.search()
rs1=re.search(r"pytho",text)
print(rs1)

#re.findall()
tx="10 10 10 11 8 8 10"
x=re.findall(r"10",tx)
print(x)

#re.finditer()-> return matched object
tx1="cat abc123 cat_bat"

rp=re.finditer(r"\w+",tx1)
print(rp)
for match in re.finditer(r"\w+",tx1):
    print(match.group(),match.start())


#re.sub()->replace text
text="python is easy"
newtext=re.sub(r"easy","awesome",tx1)
print(newtext)


#re.subn()-> replace text and count replacement
     # tx="10 10 10 11 8 8 10"
resubn=re.subn(r"10","7",tx)
print(resubn)


#re.split()-> 
t="app,is;very intersting"
print(re.split(r"[ ,;]+",t))


#re.compile() -> compile once ,use many times

pattern=re.compile(r"\d+") # "\d"= compile where pattern contain digit 0-9
print(pattern.findall("age25,roll101"))


#re.fullmatch()-> entire string must match
print(re.fullmatch(r"\d+","12345"))
#but when string mixed with letter
print(re.fullmatch(r"\d+","12345a"))

#re.escape()-> espcape special regex characters
p="price(100)"
print(re.escape(p))

```

<h2>RegEx Functions</h2>

</tr>

<tr>
    <td>finditer()</td>

</tr>

<tr>
    <td>match()</td>

</tr>

<tr>

</tr>

    <td>compile()</td>

</tr>

<tr>
    <td>escape()</td>

</tr>

</table>

<h2>RegEx Metacharacters</h2>

<table border="1" cellpadding="8" cellspacing="0">

<tr>
    <th>Character</th>
    <th>Description</th>
    <th>Example</th>
</tr>

<tr>
    <td>[]</td>
    <td>A set of characters</td>
    <td>[a-m]</td>
</tr>

<tr>
    <td>\</td>
    <td>Signals a special sequence or escapes special characters</td>
    <td>\d</td>
</tr>

<tr>
    <td>.</td>
    <td>Matches any character except newline character</td>
    <td>he..o</td>
</tr>

<tr>
    <td>^</td>
    <td>Matches the beginning of a string</td>
    <td>^hello</td>
</tr>

<tr>
    <td>$</td>
    <td>Matches the end of a string</td>
    <td>planet$</td>
</tr>

<tr>
    <td>*</td>
    <td>Matches zero or more occurrences</td>
    <td>he.*o</td>
</tr>

<tr>
    <td>+</td>
    <td>Matches one or more occurrences</td>
    <td>he.+o</td>
</tr>

<tr>
    <td>?</td>
    <td>Matches zero or one occurrence</td>
    <td>he.?o</td>
</tr>

<tr>
    <td>{}</td>
    <td>Matches exactly the specified number of occurrences</td>
    <td>he.{2}o</td>
</tr>

<tr>
    <td>|</td>
    <td>Either OR condition</td>
    <td>falls|stays</td>
</tr>

<tr>
    <td>()</td>
    <td>Captures and groups patterns</td>
    <td>(abc)</td>
</tr>

</table>

<h2>RegEx Sets</h2>

<table border="1" cellpadding="8" cellspacing="0">

<tr>
    <th>Set</th>
    <th>Description</th>
</tr>

<tr>
    <td>[arn]</td>
    <td>Matches any one character from the set: a, r, or n</td>
</tr>

<tr>
    <td>[a-n]</td>
    <td>Matches any lowercase character between a and n</td>
</tr>

<tr>
    <td>[^arn]</td>
    <td>Matches any character except a, r, and n</td>
</tr>

<tr>
    <td>[0123]</td>
    <td>Matches any digit from the set: 0, 1, 2, or 3</td>
</tr>

<tr>
    <td>[0-9]</td>
    <td>Matches any digit between 0 and 9</td>
</tr>

<tr>
    <td>[0-5][0-9]</td>
    <td>Matches any two-digit number between 00 and 59</td>
</tr>

<tr>
    <td>[a-zA-Z]</td>
    <td>Matches any alphabet character from a-z (lowercase or uppercase)</td>
</tr>

<tr>
    <td>[+]</td>
    <td>Matches the + character. Special characters lose their special meaning inside sets</td>
</tr>

<tr>
    <td>[.*+?]</td>
    <td>Matches any of the specified special characters inside the set</td>
</tr>

</table>
