# Create an iterator with stop iteration

```python
class mynum:
    def _iter_(self): # self = object of mynum class -> myclass
        self.q=1
        return self
    
    def _next_(self): 
        if self.q <=20
          x=self.q
          self.q+=1
          return x
        else: 
            raise StopIteration
          

myclass=mynum()
myIter=iter(myclass)

print(next(myIter))
print(next(myIter))
```
