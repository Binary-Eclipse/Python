Global Variable: That are created outside of a function and can used by everyone both inside and outside of function

```python
age=24

def myfunc():
    print ("i am", age ,"years old") # normal
    print(f"i am {age} years old")   # f string
    print("i am {} years old".format(age)) # format method
    
myfunc()
```

#Local variable

this can't access outside of function

```python
def myfunc():
   age=25
   print(age)
myfunc()
print(age)
```

#Global Keyword
Using Global keyword a variable can access from anywhere

```python
def global_function():
    global teeth
    teeth="white"

global_function()

print(teeth)
```
